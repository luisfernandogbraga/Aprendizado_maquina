```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('exame.csv')

colunas = {
    'Unnamed: 0': 'ID',
    'race/ethnicity': 'etnia',
    'parental level of education': 'nível educacional dos pais',
    'lunch': 'almoço',
    'test preparation course': 'curso de preparação',
    'math percentage': 'percentual de matemática',
    'reading score percentage': 'percentual de leitura',
    'writing score percentage': 'percentual de escrita',
    'sex': 'sexo'
}

df.rename(columns=colunas, inplace=True)

# Codificação one-hot das colunas 'etnia', 'almoço' e 'sexo'
sexo_encoded= pd.get_dummies(df['sexo'], prefix='sexo')
etnia_encoded= pd.get_dummies(df['etnia'], prefix='etnia')
almoco_encoded= pd.get_dummies(df['almoço'], prefix='almoço')

# Concatenar as colunas codificadas ao DataFrame
df= pd.concat([df, etnia_encoded, almoco_encoded, sexo_encoded], axis=1)

# Drop das colunas originais 'etnia', 'sexo' e 'almoço'
df= df.drop(['etnia','sexo','almoço'], axis=1)

df['nível educacional dos pais'].drop_duplicates()

# Codificação ordinal para 'nível educacional dos pais'
# colocar na orden : DO MENOR PARA O MAIOR
categorias_educacao = ["some high school", "high school", "some college", "associate's degree", "bachelor's degree", "master's degree"]
ordinal_encoder = OrdinalEncoder(categories=[categorias_educacao])
df[['nível educacional dos pais']]= ordinal_encoder.fit_transform(df[['nível educacional dos pais']])


df['curso de preparação'].drop_duplicates()
# Mapeamento para 'curso de preparação' tratando none como 0
df['curso de preparação'] = df['curso de preparação'].map({'none': 0, 'completed': 1}).fillna(0)

# Normalização para 'percentual de matemática, leitura, escrita'
min= df['percentual de matemática'].min()
max= df['percentual de matemática'].max()

df['percentual de matemática_norm']= df['percentual de matemática'].apply(lambda grade: (grade-min)/(max-min))

media= df['percentual de matemática'].mean()
desvio_padrao= df['percentual de matemática'].std()

df['percentual de matemática_padr']= df['percentual de matemática'].apply(lambda nota: (nota-media) / desvio_padrao)

media= df['percentual de matemática_padr'].mean()
desvio_padrao= df['percentual de matemática_padr'].std()
###########################################################################################                                                                         

min_1= df['percentual de leitura'].min()
max_1= df['percentual de leitura'].max()

df['percentual de leitura_norm']= df['percentual de leitura'].apply(lambda grade: (grade-min_1)/(max_1-min_1))

media_1= df['percentual de leitura'].mean()
desvio_padrao_1= df['percentual de leitura'].std()

df['percentual de leitura_padr']= df['percentual de leitura'].apply(lambda nota: (nota-media) / desvio_padrao)

media_1= df['percentual de leitura_padr'].mean()
desvio_padrao_1= df['percentual de leitura_padr'].std()
##############################################################################################

min_2= df['percentual de escrita'].min()
max_2= df['percentual de escrita'].max()

df['percentual de escrita_norm']= df['percentual de escrita'].apply(lambda grade: (grade-min_2)/(max_2-min_2))

media_2= df['percentual de escrita'].mean()
desvio_padrao_2= df['percentual de escrita'].std()

df['percentual de escrita_padr']= df['percentual de escrita'].apply(lambda nota: (nota-media) / desvio_padrao)

media_2= df['percentual de escrita_padr'].mean()
desvio_padrao_2= df['percentual de escrita_padr'].std()

df
```
